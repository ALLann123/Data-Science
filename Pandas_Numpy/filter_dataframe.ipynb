```python
import pandas as pd
```

```python
df_laptops=pd.read_csv('sample_data/laptop_price.csv', encoding='ISO-8859-1')
#lets display the loaded csv
df_laptops.head()
```

**1. Filter dataframe based on condition**

```python
#find which rows have "Apple" in the "Company" column
df_laptops['Company']=="Apple"
```

```python
#filter the df based on results
df_laptops[df_laptops['Company']=="Apple"]
```

```python
#for larger dataframes use the value_counts() method
df_laptops[df_laptops['Company']=="Apple"].value_counts('Company')
```

```python
#find which rows dont have HP inthe company column
df_laptops[df_laptops['Company']!="HP"]
```

```python
#the exact count of the number of rows without it
df_laptops[df_laptops['Company']!="HP"].value_counts('Company')
```

```python
no_hp=df_laptops[df_laptops['Company']!="HP"].value_counts('Company')
no_hp.sum()
```
