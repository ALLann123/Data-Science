```python
import pandas as pd
```

```python
#reading csv file
df_cars=pd.read_csv("sample_data/Car_sales.csv")
```

```python
#show
df_cars
```

```python
#select the columns we want
df_cars = df_cars[['Manufacturer', 'Sales_in_thousands',
                   'Vehicle_type', 'Price_in_thousands',
                   'Engine_size', 'Horsepower', 'Fuel_capacity']]
```

```python
#show dataframe
df_cars
```

```python
#instead of using value_counts we see all unique value using "nunique()"
df_cars.nunique()
```

**1 agg()**


```python
#calculate the sum of columns
df_cars.agg('sum')
```

```python
#calculate the mean of numeric colums
#---N/B select only columns that have numerical data
df_cars.select_dtypes(include='number').agg('mean')
```

```python
#calculate the mean and count
#N/B-- Select only columns that have integers
df_cars.select_dtypes(include='number').agg(['mean', 'count'])
```

```python
#Two: Mean of 'sales_in_thousands' and sum and max of 'Price_in_thousands'
df_cars.agg({'Sales_in_thousands':['sum','mean'],
             'Price_in_thousands':['sum', 'max']})
```

```python
#aggregate over the columns
df_cars[['Sales_in_thousands', 'Price_in_thousands']].agg('sum', axis=1)
```

```python
#equivalent(NaN behaves differently)
df_cars['Sales_in_thousands']+df_cars['Price_in_thousands']
```

```python
#aggregate different functions and rename the index of the resulting dataframe
df_cars.agg(x=('Sales_in_thousands', 'sum'),
            y=('Price_in_thousands', 'sum'))
```

**2 The Split-Apply-Combine Strategy**

```python
#lets see the data categories inthe columns
df_cars['Vehicle_type'].value_counts()
```

```python
#split data into separate groups
passenger_filter=df_cars['Vehicle_type']=='Passenger'
car_filter=df_cars['Vehicle_type']=='Car'
```

```python
#apply an operation
passenger_average= df_cars[passenger_filter]['Sales_in_thousands'].agg('mean')
car_average= df_cars[car_filter]['Sales_in_thousands'].agg('mean')
```

```python
#combine the results
pd.DataFrame({'Vehicle_type':['Car', 'Passenger'],
              'Sales_in_thousands': [car_average,passenger_average]}).set_index('Vehicle_type')
```

**3 groupby()**

```python
#group by vehicle_type and calculate the mean
#N/B we compute the mean using int columns only
df_cars.groupby('Vehicle_type').mean(numeric_only=True)
```

```python
#group by Manufacturer and calculate the mean
df_cars.groupby('Manufacturer').mean(numeric_only=True)
```

```python
#group by Manufacturer and then calculate the sum
df_cars.groupby('Manufacturer').sum(numeric_only=True)
```

```python
df_cars.value_counts('Manufacturer')
```

```python
#group by vehicle_type and count values
df_cars.groupby('Vehicle_type').count()
```

```python
#group by Engine_size and count values without dropping NaN values
df_cars.groupby('Engine_size', dropna=False).count()
```

**4 Groupby() and agg()**

```python
#find the minimum and maximum values on each column
df_cars.groupby('Vehicle_type').min()
```

```python
#maximum value of each column
df_cars.groupby('Vehicle_type').max()
```

```python
#get them at the same time
df_cars.groupby('Vehicle_type').agg(['min', 'max'])
```

```python
#Find the min Engine_size and max Horsepower
df_cars.groupby('Vehicle_type').agg(min_engine_size=('Engine_size', 'min'),
                                    max_horsepower=('Horsepower', 'max'))
```

```python
#calculate the sum of sales_in_thousands and mean of Price_in_thousands
df_cars.groupby('Manufacturer').agg(sum_sales=('Sales_in_thousands', 'sum'),
                                    mean_price=('Price_in_thousands', 'mean'))
```

```python
#under the hood
df_cars.groupby('Manufacturer').sum()['Sales_in_thousands']
```

```python
df_cars.groupby('Manufacturer').mean(numeric_only=True)['Price_in_thousands']
```
