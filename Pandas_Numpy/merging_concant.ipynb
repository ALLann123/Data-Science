**1.2 Concatenate Horizontally**

```python
#show shape of dataframes that we'll concatenate
df_movies.shape
```

```python
df_ratings.shape
```

```python
#find the common column and set it as an index
df_ratings.set_index('imdb_title_id', inplace=True)
df_movies.set_index('imdb_title_id', inplace=True)
```

```python
#concatenate df_movies and df_ratings on 'imdb_title_id (horizontally)
df_horizontal_concat=pd.concat([df_movies, df_ratings], axis=1)
```
