```python
import pandas as pd
```

**1 Concat()**

**1.1 Concatenate vertically**

```python
df1=pd.DataFrame({'id': ['A', 'B', 'C', 'D'],
                 'age': [30,23,25,22]})
df2=pd.DataFrame({'id': ['E','F', 'G', 'F'],
                  'age':[40,21,19,24]})
```

```python
#view
df1
```

```python
df2
```

```python
#lets concatenate the dataframes
pd.concat([df1,df2], axis=0, ignore_index=True)    #we combine the dataframes vertically
```

**1.2 Concatenate horizontally**



```python
df_1=pd.DataFrame({"id":['A', 'B', 'C', 'D'],
                   'age':[30,23,25,22]})
df_2=pd.DataFrame({'job': ['Doctor', 'Statistician', 'Account', 'Developer']})
```

```python
#View the data
df_1
```

```python
df_2
```

```python
#now lets combine the dataframes horizontally
pd.concat([df_1, df_2], axis=1)
```
