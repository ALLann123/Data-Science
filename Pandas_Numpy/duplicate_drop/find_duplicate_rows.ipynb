```python
import pandas as pd
```

```python
#load the data
df_laptops=pd.read_csv('sample_data/laptop_price.csv', encoding='ISO-8859-1')
df_laptops.head(5)
```

***1 Duplicated()***


```python
#lets identify duplicates in a column or series
df_laptops.duplicated("laptop_ID")
```

```python
#show elements in datafrae with duplicates in colun 'laptop_ID'
df_laptops[df_laptops.duplicated("laptop_ID")]   #the results show there is no duplicate column
```

```python
#look for duplicates in two or more columns
df_laptops.duplicated(["Product", "TypeName", "Inches"])
```

```python
#show all the values duplicated(except for the first -keep='first')
df_laptops[df_laptops.duplicated(["Product", "TypeName", "Inches"])]
```

```python
duplicated=df_laptops.duplicated(["Product", "TypeName", "Inches"])
```

```python
#now lets get a good view using sort
df_laptops[duplicated].sort_values(["Product", "TypeName"])
```

**1.2 Exercise**

Find the cheapest and most expensive laptop of each company using sort_value and duplicated methods

```python
#sort the dataframe aseding by company and price
#cheapest first then most expensive last
id_duplicates=df_laptops.duplicated(["Company", "Price_euros"])
df_laptops[id_duplicates].sort_values(["Price_euros"], ascending=True)
```

```python
df_laptops=df_laptops.sort_values(['Company', 'Price_euros'])
df_laptops
```

```python
#check all value categories in "Company" column
df_laptops.value_counts('Company')
```

```python
#find duplicated values in "Company" columns(defaul -> keep='first)
duplicated_first=df_laptops.duplicated('Company', keep='first')
```

```python
#show datafrae with duplicated values in "Company" column
df_laptops[duplicated_first]
```

```python
#show dataframe with non duplicated values in "company"
df_laptops[~duplicated_first][["Company", "Price_euros"]]
```

```python
#check for duplicates
df_laptops[~duplicated_first][["Company", "Price_euros"]].value_counts("Company")
```

```python
#keep is set to last
duplicated_last=df_laptops.duplicated('Company', keep='last')
```

```python
#show dataframe with non duplicated values in "company"  -Now we get laptops with most expensive
df_laptops[~duplicated_last][['Company', 'Price_euros']].sort_values("Price_euros",ascending=False)  #here we get expensive laptop per company
```

```python
#lets set the Keep=False
duplicated_false=df_laptops.duplicated('Company', keep=False)
```

```python
#show dataframe now
df_laptops[~duplicated_false]
```

```python
#all elements in company are duplicated
df_laptops.value_counts("Company")
```
