```python
import pandas as pd
```

```python
#load the data
df_laptops=pd.read_csv('sample_data/laptop_price.csv', encoding='ISO-8859-1')
df_laptops.head(5)
```

**1 drop_duplicates()**

Helps us remove duplicates in one or more columns


```python
#drop duplicates in 2 or more columns
df_laptops.drop_duplicates(['Company'])[["Company", "Price_euros"]]
```

```python
#lets verify no repeated values
df_laptops.drop_duplicates(['Company'])[["Company", "Price_euros"]].value_counts('Company')
```

```python
#sort dataframe ascending by company and price
#cheapest first and most expensive last
df_laptops=df_laptops.sort_values(['Company', 'Price_euros'])
```

```python
#cheapest: Keep='first'
df_laptops.drop_duplicates(['Company'], keep='first')[['Company', 'Price_euros']]
```

```python
#get the expensive laptops. keep='last'
df_laptops.drop_duplicates(['Company'], keep='last')[['Company', 'Price_euros']]
```

```python
#ignore_index=True the dataframe indexe's previously it had
#inplace=True updates our dataframe
df_laptops.drop_duplicates(['Company'], keep='last', inplace=True, ignore_index=True)
```

Task is to find the biggest and smallest screen size in laptops of each company using the sort_values() and duplicated() methods

```python
df_laptops=pd.read_csv('sample_data/laptop_price.csv', encoding='ISO-8859-1')
```

```python
df_laptops.head(4)
```

```python
#sort dataframe ascending by company and inches
#smallest first then biggest last
new_df_laptops=df_laptops.sort_values(['Company','Inches'])
```

```python
#smallest: keep='first'
new_df_laptops.drop_duplicates(['Company'], keep='first')[['Company', 'Inches', 'Price_euros']].sort_values('Inches')
```

```python
#Biggest: keep='last'
new_df_laptops.drop_duplicates(['Company'], keep='last')[['Company', 'Inches', 'Price_euros']].sort_values('Price_euros')
```
