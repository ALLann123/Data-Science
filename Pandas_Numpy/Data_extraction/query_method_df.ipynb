```python
import pandas as pd
```

```python
#read the csv file
df=pd.read_csv("sample_data/players_20.csv")

#set short_name to the index of our dataframe
df=df.set_index("short_name")

#select columns
df=df[["long_name", "age", "dob", "height_cm", "weight_kg", "nationality", "club"]]
```

```python
df
```

**1 Query()**

```python
#select players older than 34
df.query("age>34")

#exercise:write the equivalent boolean slicing
df[df["age"]>34]
```

```python
#select players older than 34 from italy
df.query("age>34 and nationality=='Italy'")

#equivalent bool of the method
df[(df['nationality']=="Italy")&(df['age']>34)]
```

```python
#add a not operator to the first operator
df.query("not(age>34)")   #we will get age less than 34

#lets write the bool of the query method
df[~(df["age"]>34)]     #get the age of players less than 34 using the not operator(~)
```

```python
#convert height to meters and select  those with height above 1.8m
#df.query("height_cm/100> 1.8")   #this gets the heigh in meters

df[df["height_cm"] / 100 > 1.8]
```
