```python
import pandas as pd
```

```python
#read the csv file
df=pd.read_csv("sample_data/players_20.csv")

#set short_name to the index of our dataframe
df=df.set_index("short_name")

#select columns
df=df[["long_name", "age", "dob", "height_cm", "weight_kg", "nationality", "club"]]
```

```python
#show dataframe
df
```

**1 Selecting elements with a single value**

iloc['row_position, 'column_position']

```python
#get the height of L.Messi
df.iloc[0, 3]
```

```python
#get weight of cristiano
df.iloc[1,4]
```

```python
#get all rows inside the 'height_cm'
df.iloc[:, 3]
```

```python
#GET The columns that correspond to Messi
df.iloc[0, :]
```

**2 Selecting a range of data with a slice**

```python
#get all data about Messi and Ronaldo
df.iloc[[0,1], :]
```

```python
#get the height of Messi and cr7
df.iloc[[0,1], 3]
```

```python
#get the height and weight of Messi
df.iloc[0, [3,4]]
```

```python
#get the height and weight of cr7 and Messi
df.iloc[[0,1], [3,4]]
```

**3 Selecting a range of data with a slice**

```python
#slice columns label age/club... Start is included but stop is excluded
players=[0,1]
df.iloc[players, 1:7]
```

```python
#slice of the first top10 players
columns=[1,2,3,4]

#get top 10 players
df.iloc[0:10, columns]
```

**4 Selecting with conditions**


N/B iloc[] method cannot accept boolean series but only a boolean list


```python
#select players with height above 180cm
columns=[1,2,3,4]
df['height_cm']>180  #this a bool series we convert to a list first

df.iloc[list(df['height_cm']>180), columns]
```

```python
#multiple conditions: height above 180cm and form Argentina
df.iloc[list((df['height_cm']>180) & (df['nationality']=="Argentina")), 1:7]
```
