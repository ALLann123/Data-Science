```python
import pandas as pd
```

```python
#read the csv file
df=pd.read_csv("sample_data/players_20.csv")
df
```

```python
#set short_name to the index of our dataframe
df=df.set_index("short_name")
```

```python
#show the dataframe
df
```

```python
df.head(1)
```

```python
#select columns
df=df[["long_name", "age", "dob", "height_cm", "weight_kg", "nationality", "club"]]
```

```python
#show the df
df
```

**1 Selecting with a single Value**

loc[row_label, column_label]

```python
#get all data about L.messi
df.loc["L. Messi"]
```

```python
#get the height of L.Messi
df.loc["L. Messi", "height_cm"]
```

```python
#get infor on Cr7
df.loc["Cristiano Ronaldo"]
```

```python
#get weight of CR7
df.loc["Cristiano Ronaldo", "weight_kg"]
```

```python
#get all rows inside the height_cm
df.loc[:,'height_cm']
```

```python
#select all coluns that correspong to the index 'L.Messi'
df.loc['L. Messi',:]
```

**2. Selecting with a list of values**

```python
#get all data about L.Messi and cristiano
df.loc[["L. Messi", "Cristiano Ronaldo"]]   #we add the elements(index) in our list
```

```python
#get the height of L.Messi and Cristiano Ronaldo
df.loc[["L. Messi", "Cristiano Ronaldo"], "height_cm"]
```

```python
#get the height, weight of L.Messi
df.loc["L. Messi", ["height_cm", "weight_kg"]]
```

```python
#get height anc weight of L.Messi and Cristiano Ronaldo
df.loc[["L. Messi", "Cristiano Ronaldo"], ["height_cm", "weight_kg"]]
```

**3 Selecting a range of data with a slice**

start:stop:stop
start and stop elements are included

```python
#slice column labels
players=["L. Messi", "Cristiano Ronaldo"]

#our slice goes from 'age':'club'
df.loc[players, 'age':'club']
```

```python
#slice index labels
columns=['age','dob', 'height_cm','weight_kg', 'nationality']

#get top1 and top10 player name
df.index[:10]   #the first 10 players

df.loc['L. Messi':'M. Salah', columns]

```

**4 Selecting with conditions**

```python
#one condition: select player with height above 180cm
columns=['age','dob', 'height_cm','weight_kg']

#we add condition and then the columns we want
df.loc[df['height_cm']> 180, columns]
```

```python
#Multiple conditions: select player with height above 180cm from Argentina
(df['nationality']=="Argentina") & (df['height_cm']> 180)

df.loc[(df['nationality']=="Argentina") & (df['height_cm']> 180), columns]
```

```python
#display clubs we have
df.value_counts('club')
```

```python
df[df['club']=="Manchester United"]
```

```python
#display oldest players from academy
df.loc[(df['club']=="Manchester United") & (df['age']>27), 'age':'club']
```

```python
#get our youngest players
df.loc[(df['club']=="Manchester United") & (df['age']<28), 'age':'club']
```

```python
#get youngest players and who are from england
df.loc[(df['club']=="Manchester United") & (df['age']<28) & (df['nationality']=="England"), 'age':'club']
```

```python
#lastly tallest manunited players
df.loc[(df['club']=="Manchester United") & (df['height_cm']>185), 'age':'club']
```
