```python
import pandas as pd
```

```python
df=pd.read_csv("sample_data/players_20.csv")

#set index
df.set_index('short_name', inplace=True)

#select columns
df=df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

**1 apply()**

Help us apply operations to a column

```python
#use numpy function and apply it to series
import numpy as np

#lets use the square root function
df['age'].apply(np.sqrt)
```

```python
#lets create our own functuion and apply to a dataframe
def calculate_bmi(row):
  return row['weight_kg']/(row['height_cm']/100)**2
```

```python
#apply the above function
df.apply(calculate_bmi, axis=1)
```
