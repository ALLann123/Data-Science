```python
import pandas as pd
```

```python
#read the csv file
df=pd.read_csv("sample_data/players_20.csv")

#set short_name to the index of our dataframe
df=df.set_index("short_name")

#select columns
df=df[["long_name", "age", "dob", "height_cm", "weight_kg", "nationality", "club"]]
```

```python
#show df
df
```

**1 Drop Row**

```python
#drop one row
#---1 axis parameter to drop row
df.drop('L. Messi', axis=0)     #0 rep rows and 1 rep columns
```

```python
#---2. Index parameter
df.drop(index=['L. Messi'])     #we put the index of the row we want to drop
```

```python
#drop two or more rows and update the data(inplace=True)
df.drop(['L. Messi','Cristiano Ronaldo'], axis=0, inplace=True)  #here the data is updated
```

```python
#show df
df
```

**2 Drop Columns**

```python
#drop one column
#1.  Use axis parameter
df.drop('long_name', axis=1)  #1 in axis rep column and 0 rep rows
```

```python
#2. Columns Parameter
df.drop(columns=['long_name'])
```

```python
df
```

```python
#drop column by position/index(drop last column)

df.drop(df.columns[[-1]], axis=1)   #gets the last column in our dataframe and we drop it. Use axis to select its a column
```

```python
#drop two or more columns and update the data(inplace=True)
df.drop(['long_name', 'dob'],axis=1, inplace=True)
```
