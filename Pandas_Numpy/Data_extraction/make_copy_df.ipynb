```python
import pandas as pd
```

```python
df=pd.read_csv("sample_data/players_20.csv")

#set index
df.set_index('short_name', inplace=True)

#select columns
df=df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

**1 copy()**

**1.1 deep=True**

```python
#deep=True by default
df_copy=df.copy()
```

```python
#update the value in original dataframe
df.loc['L. Messi', 'height_cm']=180
```

```python
#copy vs original
#df
df_copy
```

**1.2 deep=False**

changes inthe original dataframe affect the real data frame

```python
df_copy=df.copy(deep=False)
```

```python
#lets update height or Cr7
df.loc['Cristiano Ronaldo', 'height_cm']=153
```

```python
#copy vs original df
#df
df_copy
```
