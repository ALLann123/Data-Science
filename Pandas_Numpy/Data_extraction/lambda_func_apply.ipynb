```python
import pandas as pd
```

```python
df=pd.read_csv("sample_data/players_20.csv")

#set index
df.set_index('short_name', inplace=True)

#select columns

df=df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
df
```

**1 Lambda Function**

```python
#basic function
#sums a and b
def sum_values(a,b):
  x=a+b
  return x
```

```python
sum_values(2,3)  #call the function
```

```python
#lambda function(one liner)
sum_values_lambda=lambda a,b:a+b
```

```python
#call func
sum_values_lambda(2,3)
```

**2 Apply+Lambda Function**

```python
#use lambdafunction to convert height_cm series to meters
df['height_cm'].apply(lambda x:x/100)

#alternative
df['height_cm']/100
```

```python
#use lambda fnction to convert long_name series to uppercase
df['long_name'].apply(lambda x:x.upper())

#alternative with str attribute
df['long_name'].str.upper()
```

```python
#use lambda to get year of the dob series
df['dob']=df['dob'].astype('datetime64[ns]')  #convert column data type
```

```python
df.dtypes
```

```python
#write the function
df['dob'].apply(lambda x:x.year)

#alternative with dt attribute
df['dob'].dt.year
```

```python
#apply lambda function to dataframe in order to calculate bmi
lambda x:x['weight_kg'] / ((x['height_cm']/100)**2)

df.apply(lambda x:x['weight_kg'] / ((x['height_cm']/100)**2), axis=1)
```
