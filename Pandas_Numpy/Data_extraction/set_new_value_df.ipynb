```python
import pandas as pd
```

```python
#read the csv file
df=pd.read_csv("sample_data/players_20.csv")

#set short_name to the index of our dataframe
df=df.set_index("short_name")

#select columns
df=df[["long_name", "age", "dob", "height_cm", "weight_kg", "nationality", "club"]]
```

```python
#show dataframe loaded
df.head(10)
```

**1. Set Value to one cell**

```python
#update L.Messi Height
df.loc['L. Messi', 'height_cm']  #this displays the current height
```

```python
df.loc['L. Messi', 'height_cm']=175   #use the assignment operator to update the height
```

```python
#show the dataframe
df.head(1)
```

**2 Set value to entire column**

```python
#set height of all players to 190   ":" this selects all the columns
df.loc[:, 'height_cm']=190
```

```python
df.head(7)
```

**3 Set Value to entire row**

```python
#get all columns that correspong to player ranked last in fifa
#":"  this selects all rows,   -1 represents last number in our list
df.iloc[-1, :]
```

```python
#set null values to all columns that corresponds to player ranked last
import numpy as np
np.nan      #this introduces null values
```

```python
#lets update our data for the last player
df.iloc[-1, :]=np.nan     #we update the last player in ranking to have all there data(rows) set to null
```

```python
#show data
df.iloc[-1, :]
```

**4 Set value to Multiple Cells**

```python
#set value for all items matching the list of labels

df.loc[['L. Messi', 'Cristiano Ronaldo'], ['height_cm']]=175
```

```python
#show dataframe
df.loc[['L. Messi', 'Cristiano Ronaldo']]
```

**5 Set value for rows matching a condition**

```python
#set value for rows matching a condition
columns=['age', 'dob','height_cm','weight_kg']


df.loc[df['height_cm']>180, columns]=0  #we set all rows to 0 for players above 180
```
