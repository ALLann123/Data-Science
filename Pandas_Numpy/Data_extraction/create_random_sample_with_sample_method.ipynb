```python
import pandas as pd
```

```python
#read the csv file
df=pd.read_csv("sample_data/players_20.csv")

#set short_name to the index of our dataframe
df=df.set_index("short_name")

#select columns
df=df[["long_name", "age", "dob", "height_cm", "weight_kg", "nationality", "club"]]
```

```python
#show df
df
```

**1 Sample()**

```python
#extract 10 random elements from the "nationality" column
df['nationality'].sample(10, random_state=99)   # we get 10 random players from the dataframe

#the random_state() method controls the number of times the data generates random data
```

```python
#extract a random 20% sample of dataframe
df.sample(frac=0.2, random_state=10)      #frac stands for fraction and this provides 20% oft the data
```

```python
#Upsampling: increase the sample rate, specify the frac>1
df.sample(frac=2, replace=True, random_state=23)    #double the rows of the original dataframe
```
