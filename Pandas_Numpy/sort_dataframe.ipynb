```python
import pandas as pd
```

```python
#load the csv file
df_exams=pd.read_csv("sample_data/StudentsPerformance.csv")
df_exams.head()
```

**1 Sort a dataframe**

```python
#sort by one column
df_exams.sort_values('math score')    #by is the default parameter. The default sort algorithm is ascending
```

```python
#sort descending by one column
df_exams.sort_values('math score', ascending=False)
```

```python
#sort descending two different columns
df_exams.sort_values(['math score','reading score'], ascending=False) #add a square bracket
```

```python
df_exams    #the sort_values() method in pandas ony creates a copy of the dataframe
```

```python
df_exams.sort_values(['math score','reading score'], ascending=False, inplace=True) #inplace now updates our changes to the dataframe
```

```python
df_exams
```

```python
#sort data with text data
df_exams.sort_values("race/ethnicity", ascending=True, key=lambda col:col.str.lower())
```
