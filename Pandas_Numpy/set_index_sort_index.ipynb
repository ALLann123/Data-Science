```python
import pandas as pd
```

```python
#load the csv file
df_exams=pd.read_csv("sample_data/StudentsPerformance.csv")
df_exams.head()
```

**1. Create Index**

```python
import numpy as np
import random
```

```python
#create non repetitive values for the index
new_index=np.arange(0, 1000)
```

```python
#shuffle the indexes
random.shuffle(new_index)    #this moves the values everywhere and not in any order
```

```python
#create a new column with new index
df_exams['new_index']= new_index
```

```python
#display
df_exams
```

**2. Set Index**ðŸ˜‚

```python
#set new_index of the columns We can use it as the index of our dataframe  w
df_exams.set_index('new_index', inplace=True)   #write the column we want to set as index
```

```python
df_exams
```

```python
#now sort the dataframe by the new index column
df_exams.sort_index()    #sorts ascending by default
```

```python
#sort descending by index
df_exams.sort_index(ascending=False, inplace=True)
```
