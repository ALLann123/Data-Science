```python
import pandas as pd
```

```python
#load our datasets

df_movies=pd.read_csv('sample_data/IMDb movies.csv', low_memory=False)

df_ratings=pd.read_csv('sample_data/IMDb ratings.csv')

#now we select the columns
df_movies=df_movies[['imdb_title_id', 'title', 'year',  'genre', 'country']]

#select columns
df_ratings=df_ratings[['imdb_title_id', 'total_votes', 'mean_vote']]
```

**1 merge()**

```python
df1=pd.DataFrame({
    "id":['A','B','C','D'],
    'age':[30,23,45,20]
})
df2=pd.DataFrame({
    "id":['C','D','E','F'],
    'job':['doc','nurse','analyst','DevOps']
})
```

```python
#display
df1
```

```python
df2
```

**1.1 Inner join**

```python
df1.merge(df2, on='id',how='inner')   #gets the common columns between the data
```

```python
df_inner_join=df1.merge(df2, on='id',how='inner')
```

```python
#shape
print(df1.shape)
print(df2.shape)
print(df_inner_join.shape)
```

**1.1.1 Exercise**

```python
#merge df_movies and df_ratings (inner join)
df_movies.merge(df_ratings, on='imdb_title_id', how='inner')  #identify the common column between the two data frames
```

```python
#another way to use merge
pd.merge(df_movies, df_ratings, on="imdb_title_id")
```

**Exclusive Full Join**

Pick only values that are exclusive inthe two tables.

```python
df1.merge(df2, on='id', how='outer')
```

```python
#indicator=True
df1.merge(df2, on='id', how='outer', indicator=True)
```

```python
#query
df1.merge(df2, on='id', how='outer', indicator=True).query("_merge=='left_only' or _merge=='right_only'")
```

**Exercise**

```python
#make a copy of the df_movies dataframe
df_movies_2=df_movies.copy()
```

```python
#set the first 1000 values of 'imdb_title_id' column as 'tt1234567890'
for index in df_movies_2.index:
  #change the values of the first 1000 values
  if index<1000:
    df_movies_2.loc[index, 'imdb_title_id']='tt1234567890'

```

```python
#display the dataframe
df_movies_2.head(1000)
```

```python
#merge df_movies_2 and df_ratings(exclusive left join)
df_movies_2.merge(df_ratings, on='imdb_title_id', how='outer',
                  indicator=True).query("_merge=='left_only'")
```

```python
df_exclusive_left=df_movies_2.merge(df_ratings, on='imdb_title_id', how='outer',
                  indicator=True).query("_merge=='left_only'")
```

```python
#shape
print(df_movies_2.shape)
print(df_ratings.shape)
print(df_exclusive_left.shape)
```
