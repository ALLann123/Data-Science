```python
import pandas as pd
```

```python
df_boston=pd.read_csv('sample_data/Boston House Prices.csv')
df_boston
```

**1 Linear Regression with Scikit-learn**

```python
from sklearn import linear_model
```

**1.1 Define the dependent and independent variable**

```python
#dependent variable
y=df_boston['Value']

#independent variables
x=df_boston[['Rooms', 'Distance']]

```

**1.2 Fitting the Model**

N/B: We do not need to add the constant as sklearn adds it by default


```python
lm=linear_model.LinearRegression()

lm.fit(x, y)
```

**1.3 Predicting values**

```python
lm.predict(x)
```

```python
#if we want to select the first five values inthe array
lm.predict(x)[:5]
```

**1.4 Generate Regression Table**

```python
#r2 score
lm.score(x,y)
```

```python
#coefficient
lm.coef_
```

```python
#intercept
lm.intercept_
```
